```python
from sklearn.datasets import fetch_openml

def menu():
    print("Escolha uma opção:")
    print("1. UCI Machine Learning Repository")
    print("2. Kaggle Datasets")
    print("3. Google Dataset Search")
    print("4. OpenML")
    print("0. Sair")

def listar_conteudo_UCI():
    print("Listando conjuntos de dados disponíveis na UCI Machine Learning Repository:")
    print("1. Iris")
    print("2. Wine")
    print("3. Breast Cancer Wisconsin (Diagnostic)")
    print("4. Statlog (Heart)")
    print("5. Statlog (German Credit)")
    opcao = int(input("Digite o número do conjunto de dados desejado ou 0 para voltar: "))
    if opcao == 1:
        exibir_conteudo_UCI("iris")
    elif opcao == 2:
        exibir_conteudo_UCI("wine")
    elif opcao == 3:
        exibir_conteudo_UCI("breast-cancer-wisconsin")
    elif opcao == 4:
        exibir_conteudo_UCI("heart-statlog")
    elif opcao == 5:
        exibir_conteudo_UCI("credit-g")
    elif opcao == 0:
        menu_principal()
    else:
        print("Opção inválida.")
        listar_conteudo_UCI()

def listar_conteudo_kaggle():
    print("Listando conjuntos de dados populares do Kaggle:")
    print("1. Titanic: Machine Learning from Disaster")
    print("2. House Prices: Advanced Regression Techniques")
    print("3. Iris Species")
    opcao = int(input("Digite o número do conjunto de dados desejado ou 0 para voltar: "))
    if opcao == 1:
        exibir_conteudo_kaggle("titanic")
    elif opcao == 2:
        exibir_conteudo_kaggle("house-prices-advanced-regression-techniques")
    elif opcao == 3:
        exibir_conteudo_kaggle("iris")
    elif opcao == 0:
        menu_principal()
    else:
        print("Opção inválida.")
        listar_conteudo_kaggle()

def buscar_outros_UCI():
    print("Buscando outros conjuntos de dados na UCI Machine Learning Repository...")
    # Adicione aqui o código para buscar outros conjuntos de dados na UCI
    print("Nenhum conjunto de dados adicional encontrado.")
    voltar_menu()

def buscar_outros_kaggle():
    print("Buscando outros conjuntos de dados no Kaggle Datasets...")
    # Adicione aqui o código para buscar outros conjuntos de dados no Kaggle
    print("Nenhum conjunto de dados adicional encontrado.")
    voltar_menu()

def exibir_conteudo_UCI(nome):
    print(f"Exibindo informações sobre o conjunto de dados '{nome}':")
    metadata = fetch_openml(name=nome, data_home='data', as_frame=True)
    print(f"Nome do conjunto de dados: {metadata['target']}")
    print(f"Informações: {metadata['DESCR']}")
    opcao = input("Digite 'S' para voltar ao menu ou 'Q' para sair: ").upper()
    if opcao == 'S':
        menu_principal()
    elif opcao == 'Q':
        print("Saindo...")
    else:
        print("Opção inválida.")
        exibir_conteudo_UCI(nome)

def exibir_conteudo_kaggle(nome):
    print(f"Exibindo informações sobre o conjunto de dados '{nome}':")
    # Adicione aqui o código para exibir informações sobre o conjunto de dados do Kaggle
    print("Informações sobre o conjunto de dados não disponíveis.")
    opcao = input("Digite 'S' para voltar ao menu ou 'Q' para sair: ").upper()
    if opcao == 'S':
        menu_principal()
    elif opcao == 'Q':
        print("Saindo...")
    else:
        print("Opção inválida.")
        exibir_conteudo_kaggle(nome)

def voltar_menu():
    escolha = input("Digite 'S' para voltar ao menu ou 'Q' para sair: ").upper()
    if escolha == 'S':
        menu_principal()
    elif escolha == 'Q':
        print("Saindo...")
    else:
        print("Opção inválida.")
        voltar_menu()

def menu_principal():
    # Exibindo o menu
    menu()

    # Solicitando ao usuário que escolha uma opção
    opcao = int(input("Digite o número da opção desejada: "))

    # Listando o conteúdo disponível de acordo com a opção selecionada
    if opcao == 1:
        listar_conteudo_UCI()
    elif opcao == 2:
        listar_conteudo_kaggle()
    elif opcao == 3:
        print("Conteúdo disponível no Google Dataset Search:")
        print("Nenhum conjunto de dados disponível.")
        voltar_menu()
    elif opcao == 4:
        print("Conteúdo disponível no OpenML:")
        print("Nenhum conjunto de dados disponível.")
        voltar_menu()
    elif opcao == 0:
        print("Saindo...")
    else:
        print("Opção inválida")

# Iniciar o menu principal
menu_principal()
```
