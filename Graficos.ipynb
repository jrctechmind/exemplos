```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# **Dados para os gráficos:**

# **Gráfico de Barras:**
produtos = ["Produto A", "Produto B", "Produto C"]
vendas = [1200, 1500, 2000]

# **Gráfico de Linha:**
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho",
         "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]
trafego = [10000, 12000, 14000, 16000, 18000, 20000, 22000, 24000, 26000, 28000, 30000, 32000]

# **Gráfico de Pizza:**
despesas = ["Aluguel", "Alimentação", "Transporte", "Saúde", "Lazer"]
valores = [3000, 2000, 1500, 1000, 500]

# **Gráfico de Dispersão:**
investimento = [1000, 2000, 3000, 4000, 5000]
roi = [500, 1000, 1500, 2000, 2500]

# **Histograma:**
idades = [25, 30, 22, 27, 33, 28, 26, 29, 31, 24, 32, 23]

# **Gráfico Radar:**
categorias = ["Qualidade", "Preço", "Suporte", "Usabilidade", "Design"]
avaliacoes = [4.5, 4.2, 4.3, 4.8, 4.7]

# **Criação dos gráficos:**

plt.figure(figsize=(15, 10))  # Define o tamanho da figura

# **1. Gráfico de Barras:**
plt.subplot(2, 3, 1)
plt.bar(produtos, vendas, color=['red', 'green', 'blue'])
plt.title("Vendas por Produto")
plt.xlabel("Produto")
plt.ylabel("Vendas")
plt.grid(axis='y')

# **2. Gráfico de Linha:**
plt.subplot(2, 3, 2)
plt.plot(meses, trafego, marker='o', linestyle='-', color='blue')
plt.title("Crescimento do Tráfego do Site")
plt.xlabel("Mês")
plt.ylabel("Tráfego")
plt.grid(axis='y')

# **3. Gráfico de Pizza:**
plt.subplot(2, 3, 3)
plt.pie(valores, labels=despesas, autopct="%1.1f%%", startangle=90)
plt.title("Distribuição das Despesas")

# **4. Gráfico de Dispersão:**
plt.subplot(2, 3, 4)
plt.scatter(investimento, roi, color='red')
plt.title("Relação entre Investimento em Marketing e ROI")
plt.xlabel("Investimento em Marketing")
plt.ylabel("ROI")
plt.grid(axis='both')

# **5. Histograma:**
plt.subplot(2, 3, 5)
plt.hist(idades, bins=10, edgecolor='black')
plt.title("Distribuição de Idades dos Respondentes")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.grid(axis='y')

# **6. Gráfico Radar (using matplotlib.patches):**
theta = np.linspace(0, 2*np.pi, len(categorias), endpoint=False)
radii = avaliacoes  # Assuming ratings are between 0 and 1 for normalization

ax = plt.subplot(2, 3, 6, projection='polar')
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

# Draw polygon for the radar chart
```
